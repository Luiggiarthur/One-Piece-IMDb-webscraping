```python
from requests import get
from bs4 import BeautifulSoup
import pandas as pd 
```

```python
# Initializing the series' that the loop will 
onepiece_episodes = []

# For every season in the series
for yr in range(1999,2024):
    # Request the server the content of the web page by using get(), and store the serverâ€™s response in the variable response
    response = get('https://www.imdb.com/title/tt0388629/episodes?year=' + str(yr))

    # Parse the content of the request with BeautifulSoup
    page_html = BeautifulSoup(response.text, 'html.parser')

    # Select all the episode containers from the season page
    episode_containers = page_html.find_all('div', class_ = 'info')

    # For each episode in each season
    for episodes in episode_containers:
            # Get the info of each episode on the page
            year = yr
            episode_number = episodes.meta['content']
            title = episodes.a['title']
            airdate = episodes.find('div', class_='airdate').text.strip()
            rating = episodes.find('span', class_='ipl-rating-star__rating').text
            total_votes = episodes.find('span', class_='ipl-rating-star__total-votes').text
            desc = episodes.find('div', class_='item_description').text.strip()
            # Compiling the episode info
            episode_data = [year, episode_number, title, airdate, rating, total_votes, desc]

            # Append the episode info to the complete dataset
            onepiece_episodes.append(episode_data)
```

```python
onepiece_episodes = pd.DataFrame(onepiece_episodes, columns = ['year', 'episode_number', 'title', 'airdate', 'rating', 'total_votes', 'desc'])

onepiece_episodes
```

```python
onepiece_episodes['total_votes'] = onepiece_episodes['total_votes'].str.replace(',', '')
onepiece_episodes['total_votes'] = onepiece_episodes['total_votes'].str.replace('(', '',regex=False)
onepiece_episodes['total_votes'] = onepiece_episodes['total_votes'].str.replace(')', '',regex=False)
onepiece_episodes['total_votes'] = onepiece_episodes['total_votes'].astype(int)

onepiece_episodes['episode_number'] = onepiece_episodes['episode_number'].str.replace(',', '')
onepiece_episodes['episode_number'] = onepiece_episodes['episode_number'].astype(int)
onepiece_episodes['episode_number'] = onepiece_episodes['episode_number'].astype(int)

onepiece_episodes['rating'] = onepiece_episodes['rating'].astype(float)

onepiece_episodes['airdate'] = pd.to_datetime(onepiece_episodes['airdate'])

onepiece_episodes.info()
```

```python
onepiece_episodes.to_csv('Onepiece_episodes_IMDb_Ratings.csv',index=False)
```
